```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Make a data frame
df=pd.DataFrame({"x": np.array([0.25,0.5,0.75,1]), "3D ResNet50": np.array([47.0, 48.1, 49.5, 49.8]), "3D ResNet50 + LLaMA2 7B\nw/o rationale": np.array([33.5, 49.4, 55.5, 56.3]), "3D ResNet50 + LLaMA2 7B\nw/ rationale (ours)": np.array([63.2, 64.9, 66.1, 69])})
# Change the style of plot
plt.style.use("seaborn-darkgrid")
# Create a color palette
palette = plt.get_cmap("Set1")
# Plot multiple lines
plt.figure(figsize=(9,6))
num=0

marker_list = ["D", "^", "o"]
for idx, column in enumerate(df.drop("x", axis=1)):
    num+=1
    plt.plot(df["x"], df[column], marker=marker_list[idx], markersize=10 , color=palette(num), linewidth=3, alpha=0.9, label=column)
# Add legend
# plt.figure(figsize=(5, 3))
plt.legend(loc=2, ncol=1)
plt.ylim(30, 75) # This sets y-axis from 0% to 100%
plt.xlabel("% of training data", fontsize=22)
plt.ylabel("Accuracy (%)", fontsize=22)
# Setting x and y axis boundaries
plt.yticks(fontsize=14)
plt.xticks([0.25, 0.50, 0.75, 1.00], fontsize = 14)
# Placing the legend on the upper left
plt.legend(loc="lower center", fontsize=22.5, ncol=2)
plt.legend(bbox_to_anchor=(1.36, 1.25), fontsize=18.5, ncol=3)
# Show the graph
plt.show()
```
