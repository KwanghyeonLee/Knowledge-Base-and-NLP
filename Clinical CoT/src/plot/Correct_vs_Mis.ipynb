## Correct Diagnoses

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['Consistency', 'Correctness', 'Specificity', 'Helpfulness', 'Human-likeness']

df = {
'OPT 1.3B': [4.5, 4.35, 4.7, 4.4, 4.5],
'OPT 6.7B': [4.5, 4.35, 4.7, 4.4, 4.5],
'LLaMA2 7B': [4.5, 4.15, 4.7, 4.35, 4.4],
'ResNet + LLaMA2 7B': [4.5, 4.25, 4.5, 4.5, 4.5],
'GPT-4': [4.55, 4.65, 4.7, 4.6, 4.8],
'Ref': [4.45, 4.65, 4.75, 4.65, 4.7]
}

my_palette = ['#581845', '#900C3F', '#C70039', '#FF5733', '#FFC30F', '#B9B9B9']

x = np.arange(len(labels)) 

width = 0.2  
bar_gap_in_group = 0.03
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize=(15,5))
fig.set_facecolor('white')

for i, (model, scores) in enumerate(df.items()):
    offset = width * multiplier
    rects = ax.bar(x*1.6 + offset, scores, width - bar_gap_in_group, label=model, color=my_palette[i])
    multiplier += 1

ax.set_ylabel('Average Score', fontsize=20)
ax.set_xticks(x*1.6 + width*2.5, labels, fontsize= 26.5)
ax.legend(loc='upper left', ncols=3, fontsize=21)
ax.set_ylim(3.2, 5.3)   
plt.yticks([3.5, 3.75, 4, 4.25, 4.5, 4.75, 5], fontsize=14)

plt.show()
```
