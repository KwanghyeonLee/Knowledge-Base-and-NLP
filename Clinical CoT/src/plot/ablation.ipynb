```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

baseline = 'W/o augmentation of clinical rationales'
ours = 'Ours'
data = pd.DataFrame.from_dict({'Models':['OPT 6.7B', 'LLaMA2 7B', '3D ResNet 50\n+ OPT 6.7B', '3D ResNet 50\n+ LLaMA2 7B', 'OPT 6.7B', 'LLaMA2 7B', '3D ResNet 50\n+ OPT 6.7B', '3D ResNet 50\n+ LLaMA2 7B'],
                                "Training Data":[baseline,baseline,baseline,baseline,ours,ours,ours,ours, ],
                               "Accuracy":[33.2,34.1,41.5,56.3, 68.4,71.5,71.0,68.5]})
plt.clf() 
plt.figure(figsize=(16,6))
customPalette_1 = sns.set_palette(sns.color_palette(['#1C556C', '#F58632']))
ax = sns.barplot(x='Models',y='Accuracy',data=data,hue='Training Data', palette=customPalette_1)
ax.set_ylim([15, 85])
plt.ylabel('Accuracy (%)', fontsize=22)
plt.xlabel('', fontsize=22)
plt.yticks([20, 30, 40, 50, 60, 70, 80], fontsize=16)
plt.xticks(fontsize=22)
plt.legend(loc='upper left', fontsize=19, ncol = 2)
plt.axhline(y=68.4, color='gray', linestyle='--')
plt.show()


```
