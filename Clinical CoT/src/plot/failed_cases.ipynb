```python
import matplotlib.pyplot as plt

categories = ['Correct knowledge', 'Inappropriate expression', 'Ambiguity', 'Incorrect knowledge']
values = [62.5, 6.25, 6.25, 25]
colors = ['#A8E6CE', '#DCEDC2', '#FFD3B5', '#FF8C94']

plt.figure(figsize=(26, 8))

left = 0
for value, color, category in zip(values, colors, categories):
    plt.barh(0, value, left=left, color=color, label=category)

    plt.text(left + value / 2, 0, f"{value}%", ha='center', va='center', color='black', fontsize=30)
    left += value

plt.yticks([])  
plt.xlabel('Percentage (%)', fontsize=26)
plt.xlim(0, 100)
plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.9), fontsize=30, ncol=4) 
plt.tight_layout()
plt.show()
```
